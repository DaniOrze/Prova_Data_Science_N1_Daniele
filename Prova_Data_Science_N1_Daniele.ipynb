## Dia da semana com placares mais altos

Crie um gráfico que responda em qual dia da semana os placares foram mais altos. O conceito de "placar mais alto" é a soma dos gols feitos no jogo.

Como a quantidade de jogos não é a mesma em todos os dias da semana (finais de semana tendem a ter mais jogos que em dias da semana - o que resultaria em placares mais altos nestes dias), você deverá encontrar uma forma de equalizar esta diferença na quantidade de jogos.

A resposta deverá ser um gráfico que demonstre claramente o comparativo dos dias.

```python
# Responda aqui a etapa

# Primeiro é necessário armazenar a soma dos gols nos jogos, que seriam (gf + ga)
# Então será criada uma coluna nova para isso.
df['total_goals'] = df['gf'] + df['ga']

# Agora precisa agrupar os jogos por dia da semana
# E depois calcular a média desses placares.
goals_per_day = df.groupby('day').agg({'total_goals': 'sum', 'date': 'count'}).reset_index()
goals_per_day.columns = ['day', 'total_goals', 'total_games']
goals_per_day['avg_goals_per_game'] = goals_per_day['total_goals'] / goals_per_day['total_games']

# Ordenar os dias da semana (melhor visualização do gráfico)
days_order = ['Mon', 'Tue', 'Wed', 'Thu', 'Fri', 'Sat', 'Sun']
goals_per_day['day'] = pd.Categorical(goals_per_day['day'], categories=days_order, ordered=True)

# Agora identificar o dia com mais gols
max_goals_day = goals_per_day.loc[goals_per_day['avg_goals_per_game'].idxmax()]

# Por último, é necessário plotar o gráfico com a comparação dos dias da semana
# Será feito com base na média dos placares destacando o dia com mais gols
plt.figure(figsize=(10,6))
sns.barplot(x='day', y='avg_goals_per_game', data=goals_per_day, palette='viridis')
plt.title('Média de Gols por Jogo em Cada Dia da Semana')
plt.ylabel('Média de Gols por Jogo')
plt.xlabel('Dia da Semana')
plt.annotate(f'{max_goals_day["avg_goals_per_game"]:.2f} Gols',
             xy=(max_goals_day['day'], max_goals_day['avg_goals_per_game']),
             xytext=(max_goals_day['day'], max_goals_day['avg_goals_per_game'] + 0.1),
             arrowprops=dict(facecolor='black', shrink=0.05),
             fontsize=12, color='red', ha='center')
plt.tight_layout()
plt.show()
```

## Capitães

Quantos capitães diferentes cada equipe teve durante a temporada?

Responda em forma de um gráfico comparativo, apresentando os dados o mais claramente possível.

```python
# Responda a etapa aqui

# Para começar é necessário contar o número de capitães únicos que cada equipe teve.
# Então será agrupado por time e contado o número de capitães.
captains_per_team = df.groupby('team')['captain'].nunique().reset_index()

# Renomeando colunas para melhorar a clareza
captains_per_team.columns = ['Team', 'Different Captains']

# Ordenando os dados por número de capitães para melhorar entendimento
captains_per_team = captains_per_team.sort_values(by='Different Captains', ascending=False)

# Por último, apenas será necessário plotar o gráfico.
# Foi optado por plotar na horizontal, pois na vertical os dados ficaram muito colados
# Dificultando o entendimento do mesmo.
plt.figure(figsize=(12, 8))
sns.barplot(y='Team', x='Different Captains', data=captains_per_team, palette='magma')
plt.title('Número de Capitães Diferentes por Equipe Durante a Temporada')
plt.xlabel('Número de Capitães Diferentes')
plt.ylabel('Equipe')
plt.tight_layout()
plt.show()
```

```python
# Agora será encontrado o maior número de capitães
# E as equipes com o maior número de capitães
max_capitaes = captains_per_team['Different Captains'].max()
equipes_com_mais_capitaes = captains_per_team[captains_per_team['Different Captains'] == max_capitaes]

# Mostrando o resultado
print(f'Número máximo de capitães: {max_capitaes}')
print('Equipes com o maior número de capitães:')
print(equipes_com_mais_capitaes['Team'].tolist())
```

## Formação Tática

Qual a formação tática mais comum para cada equipe de acordo com o resultado obtido (Ex: para vitórias, a formação mais comum é X, para empates Y e para Z em derrotas).

**Importante** responda de forma que seja possível fazer comparativos entre equipes.

Responda com um gráfico.

```python
# Responda a etapa aqui

# Primeiro é necessário filtrar as colunas relevantes para a análise
# Agrupar dados por equipe, formação e resultado
# E depois, para cada time, é necessário identificar a formação mais comum por resultado
df_filtered = df[['team', 'formation', 'result']]
formation_result = df_filtered.groupby(['team', 'result', 'formation']).size().reset_index(name='count')
most_common_formation = formation_result.loc[formation_result.groupby(['team', 'result'])['count'].idxmax()]

# Agora criar gráfico combinando todos os resultados (Vitória, Derrota, Empate)
# O primeiro gráfico será mostrando as formações mais comuns por equipe e resultado
plt.figure(figsize=(12,8))
sns.set(style="whitegrid")
sns.barplot(x='team', y='count', hue='result', data=most_common_formation, dodge=True)
plt.title('Formação Tática Mais Comum por Equipe e Resultado (Vitória, Derrota, Empate)')
plt.xticks(rotation=90)
plt.ylabel('Número de Jogos')
plt.xlabel('Equipe')
plt.legend(title='Resultado')
plt.tight_layout()
plt.show()

# Agora será criado gráficos separados para derrotas (L), empates (D) e vitórias (W)
def plot_formation_by_result(result, result_label):
    result_data = most_common_formation[most_common_formation['result'] == result]
    plt.figure(figsize=(12,8))
    sns.set(style="whitegrid")
    sns.barplot(x='team', y='count', hue='formation', data=result_data, dodge=True)
    plt.title(f'Formação Tática Mais Comum para {result_label}')
    plt.xticks(rotation=90)
    plt.ylabel('Número de Jogos')
    plt.xlabel('Equipe')
    plt.legend(title='Formação')
    plt.tight_layout()
    plt.show()

plot_formation_by_result('L', 'Derrotas (L)')
plot_formation_by_result('D', 'Empates (D)')
plot_formation_by_result('W', 'Vitórias (W)')
```

## Posse de bola

Qual a média de posse de bola da equipe mandante quando ela perdeu o jogo?

Responda em forma de gráfico para poder comparar entre todas as equipes.

```python
# Responda aqui a etapa

# Primeiro é necessário filtrar os times por jogos perdidos e em casa.
home_losses = df[(df['result'] == 'L') & (df['venue'] == 'Home')]

# Agora calcular a média de posse de bola para cada time mandante
avg_possession_home_losses = home_losses.groupby('team')['poss'].mean().reset_index()

# Renomeaando colunas para melhorar a clareza
avg_possession_home_losses.columns = ['Team', 'Average Possession']

# Ordenação dos dados por média de posse de bola do maior para o menor para melhorar visualização dos dados
avg_possession_home_losses = avg_possession_home_losses.sort_values(by='Average Possession', ascending=False)

# Por último, só criar o gráfico
plt.figure(figsize=(12, 8))
sns.barplot(y='Team', x='Average Possession', data=avg_possession_home_losses, palette='coolwarm')
plt.title('Média de Posse de Bola da Equipe Mandante em Jogos Perdidos')
plt.xlabel('Média de Posse de Bola (%)')
plt.ylabel('Equipe')
plt.tight_layout()
plt.show()

# Exibição dos dados da média
print(avg_possession_home_losses)
```

## Classificação

Qual a classificação final do campeonato?

Responda em forma de tabela, ordenada do primeiro colocado ao último.
Você deverá apresentar o brasão do time (pode ser obtido online [aqui](https://logodetimes.com/premier-league/), mas automatize a busca, não salve o logo localmente), o nome, a quantidade de pontos e a posição final no campeonato.

**OBS** Vitória vale 3 pontos, empates 1 ponto e derrota 0 pontos.

```python
# Responda a etapa aqui

# Primeiro é necessário calcular os pontos de cada time de acordo com os resultados
# vitórias, derrotas ou empates.
def calcular_pontos(row):
    if row['result'] == 'W':
        return 3
    elif row['result'] == 'D':
        return 1
    else:
        return 0

df['points'] = df.apply(calcular_pontos, axis=1)
classificacao = df.groupby('team')['points'].sum().reset_index()

# Agora é necessário ordenar pela quantidade de pontos
classificacao = classificacao.sort_values(by='points', ascending=False).reset_index(drop=True)
classificacao['position'] = classificacao.index + 1

# Por último, criar uma tabela com o brasão do time, nome, quantidade de pontos e posição final.
# Função para obter o brasão do time
def obter_brasao(time):
    # Substitua espaços por '+' para a URL
    nome_formatado = time.lower().replace(" ", "-").replace("&", "and")
    url = f'https://logodetimes.com/wp-content/uploads/{nome_formatado}-football-club.png'
    return url

# Adicionar coluna de brasão
classificacao['brasao'] = classificacao['team'].apply(obter_brasao)

# Exibir a tabela
display(HTML(classificacao.to_html(escape=False, index=False)))

# Exibir a tabela formatada
for index, row in classificacao.iterrows():
    display(HTML(f"<img src='{row['brasao']}' width='50' height='50'> {row['position']} - {row['team']} - {row['points']} pontos"))
```
